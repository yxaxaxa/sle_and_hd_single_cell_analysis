```r
library(pROC)
library(ggplot2)
library(RColorBrewer)
```

```r
my.cols <- brewer.pal(5, "Set1")
```

```r
options(repr.plot.height=15)
data<-read.csv('./feature_importance_across_three_disease.csv')
colnames(data)<-c('Feature','Importance','Disease')
ggplot(data)+geom_bar(aes(y=Feature,x=Importance,fill=Disease),stat='identity',position=position_dodge(0.75),width=0.5)+theme_test()+
scale_fill_manual(values=my.cols)
```
